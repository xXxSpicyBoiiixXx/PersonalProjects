```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x_data = np.linspace(0, 10, num=100)
y_data = 10 + 3*np.sin(1.5+x_data)
```

```python
# And plot it
plt.figure(figsize=(6, 4))
plt.scatter(x_data, y_data)
plt.show()
```

```python
def test_func(x, a, b, c):
    return a + b*np.sin(c+x)

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html
params, params_covariance = curve_fit(test_func, x_data, y_data)#, p0=[1, 1, 1], bounds=((0, 0, 0), (5 5, 5))
perr = np.sqrt(np.diag(params_covariance))

print(params)
print(perr)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, test_func(x_data, params[0], params[1], params[2]), color='red', label='Fitted function')
#plt.plot(x_data, test_func(x_data, params[0], params[1], params[2]), color='red', 
#         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(params))

plt.legend(loc='best')
#plt.legend(loc='upper right')
plt.show()
```
